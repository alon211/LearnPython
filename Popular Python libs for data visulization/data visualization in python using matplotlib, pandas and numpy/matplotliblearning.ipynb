# 读取excel文件

```python
import pandas as pd

data = pd.read_excel("201708071517.xls")
print(data)
```

# 获取文件某个列将其图表化

```python
# -*- coding: utf-8 -*-
import matplotlib.pyplot as pt
import matplotlib
#防止中文乱码
from pylab import *
mpl.rcParams['font.sans-serif'] = ['SimHei'] 
data=data.head(50)
pt.plot(data["电流"],data["深度"],color="orange")
pt.scatter( data["电流"],data["深度"],color="blue")
pt.legend()
pt.xlabel("电流")
pt.ylabel("深度")
pt.show()
```

```python
import matplotlib.pyplot as pt
import pandas as pd
from pylab import *
mpl.rcParams['font.sans-serif'] = ['SimHei'] 
data = pd.read_excel("201708071517.xls")
data=data.head(50)
pt.barh(data["电流"],data["深度"],color="orange")
pt.xlabel("电流")
pt.ylabel("深度")
pt.show()
```

```python
# alpha参数表示颜色深浅，color可以是一组颜色，数据循环根据这组颜色变化
pt.bar(data["电流"],data["深度"],color=("green","blue","red","yellow"),alpha=.1)
pt.show()
```

# 创建多图表

```python
#显示子图表
#首先需要创建一个figure对象
pt.close()
fig=pt.figure()
#创建你的图标,add_subplot参数分别代表2行2列第3个
graph1=fig.add_subplot(2,2,3)
#设置title 必须要用以下方法才可以
graph1.set_title("电流 VS 深度")
graph1.plot(data["电流"],data["深度"],color=("green"))
graph1=fig.add_subplot(2,2,2)
graph1.plot(data["时间"],data["深度"],color=("green"))
pt.show()
```
